```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("D:\\OneDrive\\Desktop\\Elevate Lab Tasks\\train.csv") # or use sns.load_dataset('titanic') if seaborn is used
```

```python
# Dataset overview
df.head()

# Data types and non-null values
df.info()

# Summary statistics
df.describe(include='all')  # include='all' gives summary of categorical columns too

# Check for missing values
df.isnull().sum()

# Count unique values in categorical columns
df['Sex'].value_counts()
df['Embarked'].value_counts()
```

```python
df['Age'].hist(bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

sns.boxplot(x='Age', data=df)
```

```python
sns.countplot(x='Survived', data=df)
plt.title('Survival Counts')

sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Distribution')

sns.countplot(x='Sex', data=df)
```

```python
sns.countplot(x='Survived', hue='Sex', data=df)
```

```python
sns.countplot(x='Survived', hue='Pclass', data=df)
```

```python
sns.violinplot(x='Survived', y='Age', data=df)
```

```python
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']])
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
